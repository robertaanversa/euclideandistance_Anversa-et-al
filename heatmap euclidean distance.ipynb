```python
# Data set

df_control = pd.read_excel('control.xls', index_col=None)
df_control = df_control.set_index('x')

df_stress_minus =  pd.read_excel('stress2.xls', index_col=None)
df_stress_minus = df_stress_minus.set_index('x')

df_stress_plus =  pd.read_excel('stress.xls', index_col=None)
df_stress_plus = df_stress_plus.set_index('x')
```

```python
# Default plot control

color = sns.color_palette("Spectral", as_cmap=True)
sns.clustermap(df_control, cmap = color, vmin=0, vmax=15, method='complete',
               metric='euclidean',  linewidths=1)
```

```python
# Default plot stress minus

color = sns.color_palette("Spectral", as_cmap=True)
sns.clustermap(df_stress_minus, cmap = color, vmin=0, vmax=15, method='complete',
               metric='euclidean',  linewidths=1)
```

```python
# Default plot stress minus

color = sns.color_palette("Spectral", as_cmap=True)
sns.clustermap(df_stress_plus, cmap = color, vmin=0, vmax=15, method='complete',
               metric='euclidean',  linewidths=1)
```
